```python
def categorized_col(df,col, labels):
    
    edges=[df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]
    df[col]=pd.cut(df[col],edges , labels=labels , duplicates='drop')
    return df
    
           
```

```python
labels=['not popular', 'below average', 'average', 'popular']
categorized_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

# what is the most frequent genre of movie released on netflix

```python
df['Genre']= df['Genre'].str.split(', ')
df = df.explode( 'Genre').reset_index(drop=True)                                                                                                                
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Genre', data=df)

plt.title('Frequency of Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation= 90)
plt.show()
```

## 🎬 Most Frequent Genre — Finding
Drama, Comedy, and Action dominate the Netflix catalogue. After exploding the multi-genre column, Drama consistently appears as the most released genre, reflecting audience demand for story-driven content.

```python
# Top 10 most frequent genres on Netflix
top_genres = df['Genre'].value_counts().head(10).reset_index()
top_genres.columns = ['Genre', 'Count']

plt.figure(figsize=(12, 5))
sns.barplot(data=top_genres, x='Genre', y='Count', palette='Reds_r')
plt.title('Top 10 Most Frequent Genres on Netflix', fontsize=14)
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

print(top_genres.to_string(index=False))
```

## ⭐ Vote Average Distribution
The `Vote_Average` column was categorised into 4 ordinal buckets:
- **Not Popular** | **Below Average** | **Average** | **Popular**

This transformation converts a continuous score into a business-readable rating tier.

```python
# Distribution of Vote_Average categories
plt.figure(figsize=(8, 5))
order = ['not popular', 'below average', 'average', 'popular']
sns.countplot(x='Vote_Average', data=df, order=order, palette='coolwarm')
plt.title('Distribution of Movies by Vote Average Category')
plt.xlabel('Vote Category')
plt.ylabel('Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

print(df['Vote_Average'].value_counts())
```

## 🌍 Language Distribution
English dominates the Netflix library, but the presence of other languages highlights Netflix's push toward international content (Korean, Spanish, French, etc.).

```python
# Top 10 languages on Netflix
top_langs = df.groupby('Original_Language')['Title'].nunique().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_langs.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Top 10 Languages by Number of Unique Titles')
plt.xlabel('Language')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

print(top_langs)
```

## 📅 Releases Over the Years
Analysing release year trends shows how Netflix content production accelerated significantly after 2015, with a peak around 2019–2021.

```python
# Number of unique titles released per year
yearly = df.groupby('Release_Date')['Title'].nunique().reset_index()
yearly.columns = ['Year', 'Titles']

plt.figure(figsize=(14, 5))
sns.lineplot(data=yearly, x='Year', y='Titles', marker='o', color='crimson')
plt.title('Number of Titles Released Per Year on Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 🔥 Top 10 Most Popular Titles
Popularity scores (from TMDB) reflect combined engagement — views, searches, and ratings. Spider-Man: No Way Home leads the dataset by a significant margin.

```python
# Top 10 most popular titles (unique titles by max popularity)
top_pop = df.groupby('Title')['Popularity'].max().sort_values(ascending=False).head(10).reset_index()
top_pop.columns = ['Title', 'Popularity']

plt.figure(figsize=(12, 5))
sns.barplot(data=top_pop, x='Popularity', y='Title', palette='magma')
plt.title('Top 10 Most Popular Titles on Netflix')
plt.xlabel('Popularity Score')
plt.ylabel('Title')
plt.grid(axis='x')
plt.tight_layout()
plt.show()

print(top_pop.to_string(index=False))
```

## 🔗 Correlation — Popularity vs Vote Count
Do more votes mean more popular? Checking the relationship between `Popularity` and `Vote_Count` reveals whether crowd engagement aligns with visibility.

```python
# Scatter: Popularity vs Vote_Count
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='Vote_Count', y='Popularity', alpha=0.4, color='darkorange')
plt.title('Popularity vs Vote Count')
plt.xlabel('Vote Count')
plt.ylabel('Popularity Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Numeric correlation
print("Correlation between Popularity and Vote_Count:")
print(round(df[['Popularity','Vote_Count']].corr().iloc[0,1], 4))
```
